```python
import pandas as pd
import nltk
from nltk import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
data = pd.read_csv('https://raw.githubusercontent.com/pycaret/pycaret/master/datasets/amazon.csv')
data
```

```python
sentences = sent_tokenize(data.reviewText[17])
words = word_tokenize(sentences[1])
stop_words = stopwords.words('english')
words = [w for w in words if not w in stop_words and w.isalpha()]
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

textoWC = ' '.join(words)
wordcloud = WordCloud(
    width = 250,
    height = 250,
    background_color = 'white',
    stopwords = STOPWORDS,
    relative_scaling = 0,
    min_font_size = 8
).generate(textoWC)

plt.figure(figsize = (6,6), facecolor = None)
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
```
